```python
import matplotlib.pyplot as plt
import numpy as np

countries = ["India", "China", "US", "UK", "Australia"]
total_students = [248_000_000, 291_000_000, 50_000_000, 10_000_000, 3_000_000]
struggling_students = [25_000_000, 22_000_000, 10_000_000, 3_000_000, 500_000]

fig, ax = plt.subplots(figsize=(10,6))

bar_width = 0.35
x = np.arange(len(countries))

bars_total = ax.bar(x - bar_width/2, total_students, bar_width,
                    label='Total Students', color='skyblue',
                    edgecolor='darkblue', linewidth=1.5)

bars_struggling = ax.bar(x + bar_width/2, struggling_students, bar_width,
                         label='Struggling Students', color='salmon',
                         edgecolor='red', linewidth=1.5)

ax.set_xlabel('Country')
ax.set_ylabel('Number of Students')
ax.set_title('Students vs Students Struggling in Math')
ax.set_xticks(x)
ax.set_xticklabels(countries)
ax.legend()

plt.show()
```
