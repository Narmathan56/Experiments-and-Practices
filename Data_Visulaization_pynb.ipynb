```python
Data={"Salary":[2000,1500,5000,10000,3000,4000]*4}
Df=pd.DataFrame(Data)
Df.head()
```

```python
Df.shape
```

```python
plt.plot(Df,color="red", marker="o")
plt.grid()

plt.show()
```

```python
Df['dept']=['IT','HR','Finance','HR','Finance','IT','IT','HR','Finance','HR','HR','HR']*2
```

```python
Df.head()
```

```python
Df.info()
```

```python
count=Df['dept'].value_counts()
count.head()
```

```python
##pie chart
plt.pie(count, labels=count.index, autopct='%1.2f',explode=[0,0.01,0.2])
plt.show()
```

```python
# Count plot'
plt.bar(count.index,count,color=['red','yellow','green'])
plt.show()
```
